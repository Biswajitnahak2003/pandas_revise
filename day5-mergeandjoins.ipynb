```python
import pandas as pd

employees = pd.DataFrame({
    "EmpID": [101, 102, 103, 104],
    "Name": ["Alice", "Bob", "Charlie", "David"],
    "DeptID": [1, 2, 1, 3]
})

print(employees)
```

```python
departments = pd.DataFrame({
    "DeptID": [1, 2, 4],
    "Department": ["IT", "HR", "Marketing"]
})

print(departments)
```

```python
## merge
pd.merge(employees, departments, on = "DeptID")
```

```python
result = pd.merge(
    employees,
    departments,
    on="DeptID",
    how="inner"
)

print(result)
```

```python
## inner join (default)
result = pd.merge(
    employees,
    departments,
    on = "DeptID",
    how="left"
)
print(result)
```

```python
## right join
result = pd.merge(
    employees,
    departments,
    on = "DeptID",
    how="right"
)
print(result)
```

```python
## outer join (keeps everything)
result = pd.merge(
    employees,
    departments,
    on = "DeptID",
    how="outer"
)
print(result)
```

```python
## merge on different column name use left_on and right_on
pd.merge(
    employees,
    departments,
    left_on="DeptID",
    right_on="DeptID"
)
```

```python
## merge on multiple columns (both column names should match)
employees = pd.DataFrame({
    "EmpID": [101, 102, 103, 104],
    "Name": ["Alice", "Bob", "Charlie", "David"],
    "city": ["Delhi", "Banglore", "Hyderabad", "Bhubaneswar"],
    "DeptID": [1, 2, 1, 3]
})
departments = pd.DataFrame({
    "DeptID": [1, 2, 4],
    "Department": ["IT", "HR", "Marketing"],
    "city": ["Delhi", "Banglore", "Pune"]
})
merged_df = pd.merge(
    employees,
    departments,
    on=["DeptID", "city"]
)
print(merged_df)
```

```python
## duplicate column names
pd.merge(
    employees,
    departments,
    on = "DeptID",
    suffixes=("_emp","_cit")
)
```

```python
## concatenate
df1 = pd.DataFrame({
    "Name": ["Alice", "Bob"]
})

df2 = pd.DataFrame({
    "Name": ["Charlie", "David"]
})
pd.concat([df1, df2])
```

```python
pd.concat(
    [df1, df2],
    ignore_index=True
)
```

```python
## horizontal concat
pd.concat(
    [df1, df2],
    axis=1
)
```

```python
## join
employee = employees.set_index("DeptID")
salary = departments.set_index("DeptID")

employee.join(salary).reset_index()
```

```python
## merge indicator
pd.merge(
    employees,
    departments,
    on="DeptID",
    how="outer",
    indicator=True
)
```

```python
## validate merge
pd.merge(
    employees,
    departments,
    on="DeptID",
    validate="many_to_one"
)
```
