```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    "Employee": ["Alice","Bob","Charlie","David","Eva"],
    "Salary":[30000,55000,70000,45000,80000],
    "Age":[22,28,35,30,26],
    "JoinDate":[
        "2023-01-15",
        "2022-08-20",
        "2021-03-10",
        "2023-11-01",
        "2020-07-18"
    ]
})

print(df)
```

```python
df["JoinDate"].dtype
```

```python
df["JoinDate"] = pd.to_datetime(df["JoinDate"])
```

```python
df.dtypes
```

```python
df["Year"] = df["JoinDate"].dt.year    ## adding new column as year
```

```python
df
```

```python
df["Month"] = df["JoinDate"].dt.month    ## adding new column as month
df["Day"] = df["JoinDate"].dt.day    ## adding new column as day
df["Weekday"] = df["JoinDate"].dt.day_name()    ## adding new column as day name
df
```

```python
df[df["JoinDate"] > "2023-01-01"]
```

```python
### apply()
df["Bonus"] = df["Salary"].apply(
    lambda x: x*0.1
)
df
```

```python
## map()  works on series only
city = pd.Series([
    "Delhi",
    "Mumbai",
    "Kolkata"
])
```

```python
city.map({
    "Delhi":"DL",
    "Mumbai":"MH",
    "Kolkata":"WB"
})
```

```python
## apply() on Entire Rows
def calculate_tax(row):

    if row["Salary"] > 60000:
        return row["Salary"] * 0.30

    return row["Salary"] * 0.10

df["Tax"] = df.apply(
    calculate_tax,
    axis=1
)
```

```python
df
```

```python
## replace()
df["Employee"] = df["Employee"].replace(
    "Bob",
    "John"
)
df
```

```python
## cut()
df["SalaryBand"] = pd.cut(
    df["Salary"],
    bins=[0,40000,60000,100000],
    labels=["Low","Medium","High"]
)
df
```

```python
## qcut()    divides equally
pd.qcut(
    df["Salary"],
    q=4
)
df
```

```python
## where()
df["Salary"] = df["Salary"].where(
    df["Salary"] >= 50000
)
df
```

```python
## np.select()    Multiple conditions
conditions = [
    df["Salary"] >= 70000,
    df["Salary"] >= 50000
]

choices = [
    "High",
    "Medium"
]

df["Level"] = np.select(
    conditions,
    choices,
    default="Low"
)
df
```

```python
## assign()
df = df.assign(
    Bonus=df["Salary"] * 0.1
)
df
```
